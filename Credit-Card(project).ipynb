```python
from pycaret.datasets import get_data
dataset = get_data('credit')
```

```python
#check the shape of data
dataset.shape
```

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

## Setting up Enviroment in PyCaret()

```python
from pycaret.classification import *
```

```python
exp_clf = setup(data = data, target = 'default', session_id=123)
```

## Comparing All Models

```python
best_model = compare_models()
```

```python
print(best_model)
```

```python
#classifiers available in model library of PyCaret
models()
```

## Decision Tree Classifier

```python
dt = create_model('dt')
```

```python
#trained model object is stored in the variable 'dt'. 
print(dt)
```

## K Neighbors Classifier

```python
knn = create_model('knn',fold=5)
```

## Random Forest Classifier

```python
rf = create_model('rf')
```

```python
tuned_rf = tune_model(rf)
```

## Plot Model  

```python
#AUC PLOT
plot_model(tuned_rf, plot='auc')
```

```python
# Precision-Recall Curve
plot_model(tuned_rf , plot='pr')
```

```python
# Feature Importance Curve
plot_model(tuned_rf , plot='feature')
```

```python
#Confusion Matrix
plot_model(tuned_rf , plot='confusion_matrix')
```

```python
#evaluate_model(tuned_rf)
#displays a user interface for all of the available plots for a given model.
```

## Predict Model

```python
predict_model(tuned_rf);
```

## Finalize Model For Deployment

```python
final_rf = finalize_model(tuned_rf)
```

```python
# final random forest model parameters for deployment
print(final_rf)
```

```python
predict_model(final_rf);
```

## predict on unseen data

```python
unseen_predictions = predict_model(final_rf, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions['default'], unseen_predictions['Label'], metric = 'Accuracy')
```
